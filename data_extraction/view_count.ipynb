```python
import pandas as pd
```

```python
translated = pd.read_csv("not_translated_filtered.csv")
translated.head()
```

```python
import requests
from urllib.parse import quote, urlencode
import time
from datetime import datetime
import sys
import numpy as np

views_df = pd.DataFrame()
artcount = translated.shape[0]
for index, row in translated.iterrows():
    try:
        if (index + 1) % 1000 == 0:
            print("{0} out of {1}".format(index, artcount))
        url = "https://wikimedia.org/api/rest_v1/metrics/pageviews/per-article/uk.wikipedia.org/all-access/all-agents/{0}/monthly/20180101/20180718"
        url = url.format(quote(row.title, safe=''))
        res = requests.get(url)

        views = 0
    
        json = res.json()
        items = json["items"]

        views = np.zeros(7)
        for i in items:
            date = datetime.strptime(i["timestamp"], '%Y%m%d%H')
            views[date.month - 1] = i["views"]
        views_df[index] = views
    except:
        print("Unexpected error:", sys.exc_info()[0])
        print("Index: {0}, no data for title {1}\n url: {2}".format(index, row.title, url))
```

```python
views_df.shape
```

```python
views_df.to_csv('translated_views_from_2016_to_2017_without_page_id_and_title_only_16k.csv', index=False)
```

```python
new_data = pd.DataFrame()
new_data["page_id"] = translated["page_id"]
new_data["title"] = translated["title"]
new_data = pd.concat([new_data, views_df.T], axis=1, sort=False)
new_data.to_csv("not_translated_views_from_2018.csv")
```
